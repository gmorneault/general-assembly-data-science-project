```python
%run import_clean_data.ipynb
%matplotlib inline
import matplotlib.pyplot as plt
```

## Histograms

```python
fig,axes = plt.subplots(nrows=2, ncols=4)
fig.set_size_inches(25,10)

axes[0,0].hist(loanStats.dti.dropna())
axes[0,0].set_title('Debt to Income')

axes[1,0].hist(loanStats.int_rate.dropna())
axes[1,0].set_title('Interest Rate')

axes[0,1].hist(loanStats.fico_range_high.dropna())
axes[0,1].set_title('FICO Score')

axes[1,1].hist(loanStats.emp_length_num.dropna())
axes[1,1].set_title('Employment Length')

axes[0,2].hist(loanStats.funded_amnt[loanStats.funded_amnt < 50000], bins=20)
axes[0,2].set_title('Funded Amount')

axes[1,2].hist(loanStats.annual_inc[loanStats.annual_inc < 300000], bins=20)
axes[1,2].set_title('Annual Income')

axes[0,3].hist(loanStats.total_acc.dropna())
axes[0,3].set_title('Total Accounts')

axes[1,3].hist(loanStats.revol_util[loanStats.revol_util < 200])
axes[1,3].set_title('Revolving Utilization Rate')
```

## FICO vs Interest Rate by Grade

```python
fig = plt.figure(figsize=(15,10))
ax1 = fig.add_subplot(1,1,1)

colors = plt.get_cmap('RdYlBu')
grade_colors = {
    'A': colors(1.0),
    'B': colors(0.9),
    'C': colors(0.7),
    'D': colors(0.5),
    'E': colors(0.3),
    'F': colors(0.1),
    'G': colors(0.0),
}

for c_g in grade_colors:
    ax1.scatter(loanStats[loanStats.grade==c_g].fico_range_low, loanStats[loanStats.grade==c_g].int_rate, c=grade_colors[c_g], label=c_g)

ax1.legend(loc='best')
ax1.set_title('Interest Rate vs FICO Score by Grade')
ax1.set_xlabel('FICO Score')
ax1.set_ylabel('Interest Rate')
```

# Stacked Bar Plots

```python
fig,axes = plt.subplots(nrows=2, ncols=2)
fig.set_size_inches(25,15)

home_own_filter = loanStats[loanStats.home_ownership.apply(lambda x: x in ['MORTGAGE', 'RENT', 'OWN'])]
home_grade_count = pd.crosstab(home_own_filter.grade, home_own_filter.home_ownership)
home_grade_pct = home_grade_count.div(home_grade_count.sum(1).astype(float), axis=0)
home_grade_pct.plot(kind='bar', stacked=True, ax=axes[0,0])
axes[0,0].set_title('Percent of Home Ownership by Grade')
axes[0,0].legend(loc='upper right')

bad_grade_count = pd.crosstab(loanStats.grade, loanStats.is_bad)
bad_grade_pct = bad_grade_count.div(bad_grade_count.sum(1).astype(float), axis=0)
bad_grade_pct.plot(kind='bar', stacked=True, ax=axes[0,1])
axes[0,1].set_title('Percent of Bad Accounts by Grade')
axes[0,1].legend(loc='upper right')

bad_vintage_count = pd.crosstab(loanStats.vintage, loanStats.is_bad)
bad_vintage_pct = bad_vintage_count.div(bad_vintage_count.sum(1).astype(float), axis=0)
bad_vintage_pct.plot(kind='bar', stacked=True, ax=axes[1,0])
axes[1,0].set_title('Percent of Bad Accounts by Vintage')
axes[1,0].legend(loc='lower right')

inc_v_grade_count = pd.crosstab(loanStats.grade, loanStats.is_inc_v)
inc_v_grade_pct = inc_v_grade_count.div(inc_v_grade_count.sum(axis=1).astype(float), axis=0)
inc_v_grade_pct.plot(kind='bar', stacked=True, ax=axes[1,1])
axes[1,1].set_title('Income Verification by Grade')
axes[1,1].legend(loc='upper right')
```
